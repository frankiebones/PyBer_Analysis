```python
import matplotlib.pyplot as plt
```

# Line Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```


# Object-Oriented Interface Method  
> ax.plot()

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2)
ax.legend(["Boston"], loc='upper left')
ax.set_title("PyBer Rides")
ax.set_xlabel("Month")
ax.set_ylabel("Fare in $")
ax.set_ylim(5, 50)
ax.grid()
```

## further broken down example of  
> fig, ax = plt.subplots()

```python
# Create the plot with ax.plt()
fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# create plot and add label for legend
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')

# create labels for x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# create a title
plt.title("PyBer Fare by Month")

# set the y axis limit between 0 and 45
plt.ylim(0, 45)

# add legend
plt.legend()
```

# Bar Charts  
> **vertical**  
>
> plt.bar( ) 

```python
# create the plot
plt.bar(x_axis, y_axis, color='green', label='Boston')

# create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# create title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()
```

> **horizontal**  
>  
> plt.barh( )  
  
>  **get current axes**
>
> .gca( )

```python
# Create the plot
plt.barh(x_axis, y_axis, color='magenta', label='Boston')

# invert the y axis labels
plt.gca().invert_yaxis()

plt.title("PyBer Fare per Month")
plt.xlabel("Fare in Dollars ($)")
plt.ylabel("Month")
plt.legend()
```

# Object-Oriented Interface Method for Bar Charts

> ax.bar( )

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()

ax.bar(x_axis, y_axis, color="xkcd:sky blue", label='New York')
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()

ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_title("PyBer")
ax.set_xlabel("Month")
ax.set_ylabel("Fare in $")
ax.invert_yaxis()
ax.legend()
```

# Scatter Plot
> plt.plot('o')

```python
# create scatter plot by using 'o' in plt.plot('o')
plt.plot(x_axis, y_axis, 'o')
```

> plt.scatter( )

```python
# create scatter plot by using plt.scatter()
plt.scatter(y_axis, x_axis, c='red', s=y_axis)
plt.gca().invert_yaxis()
```

# Bubble Chart

```python
# increase the size of the entire list to make the base size of bubble larger
# for loop method
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

## List Comprehensions  

-faster than for/while loops  
-optimized for Pythion to spot a predictable pattern  
>new_list = [expression for item in list if conditional]

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

# Object-Oriented Scatter Plot

```python
fig, ax = plt.subplots()

ax.scatter(x_axis, y_axis, s=y_axis, c='xkcd:sky blue', linewidth=2, alpha=0.75, edgecolor="k")
```

# Pie Charts
> plt.pie( )

```python
plt.pie(y_axis, labels=x_axis)
```

## explode 
will offset the indicated wedge by a fraction of the radius  
0 = zero distance from center of pie  
1 = completely outside the diameter of pie  
length of the supplied tuple for explode must match the amount of wedges

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

## assigning colors

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "plum", "lightblue", "green", "yellowgreen", "greenyellow", "wheat", "orange", "gold", "lightcoral", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

# Object-Oriented Pie Chart
